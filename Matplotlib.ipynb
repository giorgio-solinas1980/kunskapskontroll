### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

A Figure is an Artist that hols all the elements used to create a plot. Artists like Axes, Axis, Titles, etc are all contained in a Figure. It is a sort of main canvas for the plot. A Figure can have 1 or more Axes. An Axes is the area where the plot/subplots will be displayed: it includes the Artists needed to configurate our data system (Axis), the plotted data, ticks, labels and titles. We call the methods to create the plotting directly from the Axes. The Axis are the reference for the plot setting the scale and the limit, the ticks. An Artist is basically anything we see needed to gerenare our plot! 

### When plotting in Matplotlib, what is the expected input data type?

Numpy arrays or objectes that can be converted in numpy arrays!

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(-4,4)
y=x**2
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('My first plot')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(100)
random_data_y = np.random.randn(100)
x = np.linspace(-2, 2, 100)
y = x**2
fig, ax = plt.subplots(1,2)
ax[0].plot(x,y,"b--", label="y=x")
ax[1].scatter(random_data_x, random_data_y, c="red", label="random")
ax[0].set_title("Random numbers vs parabola")
ax[0].legend()
ax[1].legend()
# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
fig, bx = plt.subplots()
bx.bar(names, values, color=["green","red","orange","yellow", "lime" ])
bx.set_title("Amount of fruits")
```
